```python
df=pd.read_csv('dataset1.csv',encoding = "ISO-8859-1")
df
```

```python
a=df.drop(columns=['address','description','page_url','rent','sale']).replace('--',np.nan).replace('No Data',np.nan).dropna()
a['zipcode']=a['zipcode'].apply(int).apply(str)
a.replace('nan',np.nan)
a['year built']=a['year built'].apply(int)
a['bed']=a['bed'].apply(int)
a['bath']=a['bath'].apply(int)
a['size']=a['size'].apply(int)
a.info()
```

```python
a.to_csv('data for visual.csv',encoding='utf-8',index=False)
price_group=a[['zipcode','price']].groupby('zipcode').mean()
ddf=pd.DataFrame(price_group).reset_index()
#ddf.reset_index(level=0, inplace=True)

ddf
```

```python
import folium
#Center the map at Times Square
m = folium.Map(location = [40.7589,-73.9851],zoom_start=13)

m.choropleth(geo_data='zipcode.geojson.json', data=ddf,  # geo_data='zipcode.geojson' take the object, api
             columns=['zipcode','price'],
             key_on='feature.properties.postalCode',
             fill_color='RdYlBu', fill_opacity=0.7, line_opacity=0.8,
             legend_name='Average House Price groupby Zipcode')
folium.LayerControl().add_to(m)
m
```
