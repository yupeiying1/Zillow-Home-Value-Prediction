```python
import pandas as pd
sdf = pd.read_csv('~/Desktop/da_project/csv_file/data_total_complaint.csv')
```

```python
len(sdf)
```

```python
#only need crime complaint data of Manhattan
import pandas as pd
df = sdf
zip_groups = df.groupby("zipcode")
# print(zip_groups.size())
sizes = pd.DataFrame(zip_groups.size())
sizes.rename(columns={0:"size"},inplace=True)
sizes.reset_index(level=0, inplace=True)

sizes = sizes[(sizes['zipcode'] <= 10280)]
# #               * (sizes['zipcode'] <= 10044)
# #               or sizes['zipcode']==10065 or sizes['zipcode']==10075 
# #               or sizes['zipcode']==10128 or sizes['zipcode']==10280]
#make up lose data area based on https://www1.nyc.gov/site/nypd/news/pr1107/nypd-sees-continued-decline-crime-october-2018#/0
df_lost =pd.DataFrame({"zipcode":[10036], "size":[6536]}) 
sizes = sizes.append(df_lost,'sort=False')
sizes.head()
```

```python
#import zipcode geojson data to prepare heatmap making 
# '~/Desktop/da_project/csv_file/'

filename = '~/Desktop/da_project/csv_file/zipcode.geojson'
sizes['Zip'] = sizes['zipcode'].astype('str')
# sizes.to_csv( '/Users/jianxingwan/Desktop/size.csv')
```

```python
#crome data heatmap in the total Manahattan area
import folium
import json
import geojsonio
#Center the map at Times Square
m = folium.Map(location = [40.7589,-73.9851],zoom_start=12)

m.choropleth(geo_data='/Users/jianxingwan/Desktop/da_project/zipcode.geojson', data=sizes,
             columns=[ 'Zip','size'],
             key_on='feature.properties.postalCode',
             fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.8,
             legend_name='Distribution of Safety issues')
folium.LayerControl().add_to(m)
m
```
